```python
import pandas as pd
import numpy as np
import gzip as gz
```

### Compress the 2022 data

```python
# Load df
proba_df = pd.read_csv("./data/probas_2022.csv", index_col=0)
```

```python
# Chunk and Compress
N = 10
n = int(proba_df.shape[1]/N)

for i in range(N):
    filename = "./data/compr_probas_2022_%02d.csv.gz" % i
    proba_df.iloc[:, (i*n):(n+i*n)].to_csv(filename, compression="gzip")
```
